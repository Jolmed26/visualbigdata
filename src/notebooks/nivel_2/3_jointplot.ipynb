# Joint Plot
Es un gráfico compuesto por dos tipos de gráficos diferentes: Es un Scatter Plot que al mismo tiempo muestra histogramas (o KDE) con la distribución de cada variable. Para este gráfico, ambas variables deben ser numéricas.    
**Eje X**: Variable numérica  
**Eje Y**: Variable numérica  



## En esta clase
### Datos
Información sobre especies de pinguinos
### Objetivos
1. Encontrar alguna correlación entre las características de los pinguinos y al mismo tiempo ver la distribución de las variables

## Referencias
- https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df
```

```python
# Eliminamos filas NaNs
df = df.dropna()
df
```

```python
# Jointplot Flipper lenght vs Culmen Lenght 
# https://seaborn.pydata.org/generated/seaborn.jointplot.html
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    #kind="reg"
)
```

```python
# Jointplot Flipper lenght vs Culmen Lenght 
# https://seaborn.pydata.org/generated/seaborn.jointplot.html
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    hue='species'
)
```
