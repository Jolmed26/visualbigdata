# Circle Packing (*Clase en Preparación*)
Es hermano del TreeMap. Tiene la misma funcionalidad, las mismas ventajas y desventajas. Pero en ves de cuadrados, se utilizan círculos para agrupar las diferentes categorias. Es un gráfico jerárquico sin ejes. El gráfico realiza un "groupby" visual de varios niveles y muestra la variable numérica agregada resultante de cada agrupación diferente como un círculo. El tamaño de este círculo es proporcional al valor de esta variable numérica agregada. Los niveles se agrupan en círculos uno dentro del otro, siendo el primer nivel el círculo mas externo.




### Cuando utilizar
- Mi dataset tiene variables categóricas que codifican grupos de una manera jerárquico y quiero conocer las diferencias entre las **magnitudes** de los valores de variables numéricas agregadas dentro de estos diferentes grupos.

### Cuando no utilizar
- Si mis variables categóricas jerárquicas a agrupar son temporales, por ejemplo, años - meses - dias, no es común hacer un treemap para estos casos. En estos casos sería preferible un gráfico de líneas para ver las diferencias a través del tiempo.
- Si quiero visualizar las diferencias de las **proporciones** entre los grupos de un mismo nivel. En este caso debemos utilizar un Sunburst.


### Consejos
- TreeMap o Circle Packing? El Circle Packing es mas atractivo visualmente. Pero no es eficiente con el espacio. Por lo que, si debemos presentar la visualización en algun documento formal, se prefiere el TreeMap.
- Es común utilizar colores en el gráfico para distinguir cada categoría en el primer nivel. También se pueden utilizar los colores para codificar la variable numérica o una segunda variable numérica relacionada a los grupos.
- Podría ser una alternativa a un Gráfico de Barras con muchas categorias. Sin embargo, perdemos presición. Aún mas si las magnitudes son similares entre sí.
